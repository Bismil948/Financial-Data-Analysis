```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,5)

df = pd.read_csv('all_stocks_5yr.csv')
df['date'] = pd.to_datetime(df['date'])

company = 'AAL'
df = df[df['Name'] == company]

print(df.head())
print(df.info())
print(df.describe())

df = df.dropna()

plt.figure()
sns.histplot(df['close'], kde=True, color='blue')
plt.title(f"Distribution of Closing Prices for {company}")
plt.xlabel("Closing Price")
plt.ylabel("Frequency")
plt.show()

plt.figure()
sns.lineplot(data=df, x='date', y='close', marker='o')
plt.title(f"Closing Price Trend Over Time for {company}")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.xticks(rotation=45)
plt.show()

plt.figure()
sns.lineplot(data=df, x='date', y='volume', color='green')
plt.title(f"Volume Traded Over Time for {company}")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.xticks(rotation=45)
plt.show()

plt.figure()
sns.heatmap(df[['open','high','low','close','volume']].corr(), annot=True, cmap='coolwarm')
plt.title(f"Correlation Between Numerical Features for {company}")
plt.show()

plt.figure()
sns.scatterplot(data=df, x='open', y='close', hue='volume', palette='viridis')
plt.title(f"Open vs Close Prices with Volume for {company}")
plt.xlabel("Open Price")
plt.ylabel("Close Price")
plt.show()

insights = f"""
Key Insights for {company}:
1. Closing prices trend can be observed over time with fluctuations.
2. Strong correlation exists between Open and Close prices.
3. Trading volume shows spikes on certain days.
4. Distribution of closing prices shows central tendency and spread.
5. Scatter plot indicates Open and Close prices are closely related.
"""
print(insights)
```
